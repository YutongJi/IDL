```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import signal

omega = 1.0
gamma = 0.06
A = 1.0
w = 1.0
fs = 10e2
num_sims = 5 ### display five runs
t_init = 0
t_end  = 700
n = 1000### Compute 1000 grid points
dt     = math.pi *2/(n*min(omega,w))
x_init = 0
z_init = 1.0
ts    = np.arange(t_init, t_end, dt)
N = ts.size

def mu(x, z, t):
    return -(gamma*z + omega*x)
    
def Cos(t): 
    return A*math.cos(w*t)

xs    = np.zeros(N)
zs    = np.zeros(N)

xs[0] = x_init
zs[0] = z_init

for i in range(1, ts.size):
    t = (i-1) * dt
    x = xs[i-1]
    z = zs[i-1]
    zs[i] = z + mu(x, z, t) * dt + Cos(t)
    xs[i] = x + z*dt
ts1 = ts [30000:]
xs1 = xs [30000:]
ps = np.abs(np.fft.fft(xs))**2

time_step = 1 / 30
freqs = np.fft.fftfreq(xs.size, time_step)
idx = np.argsort(freqs)
plt.plot(ts,xs)
#plt.plot(freqs[idx], ps[idx])
plt.show()
#print(xs1.size)
```
