```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

fs = 10e2
num_sims = 5 ### display five runs
t_init = 3
t_end  = 70
N      = 1000 ### Compute 1000 grid points
dt     = float(t_end - t_init) / N 
y_init = 0

c_theta = 0.7
c_mu    = 1.5
c_sigma = 0.06
def mu(y, t):
    """Implement the Ornstein–Uhlenbeck mu.""" ## = \theta (\mu-Y_t)
    return c_theta * (c_mu - y)

def sigma(y, t): 
   ## = \sigma
    return c_sigma
    
def dW(delta_t): 
    """Sample a random number at each call."""
    return np.random.normal(loc = 0.0, scale = np.sqrt(delta_t))

ts    = np.arange(t_init, t_end, dt)
ys    = np.zeros(N)

ys[0] = y_init

for _ in range(num_sims):
    for i in range(1, ts.size):
       t = (i-1) * dt
       y = ys[i-1]
       ys[i] = y + mu(y, t) * dt + sigma(y, t) * dW(dt)
    plt.plot(ts, ys)
    f, Pxx_den = signal.periodogram(xs, fs)
    plt.semilogy(f, Pxx_den)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

omega = 1.0
gamma = 0.06
A = 1.0
w = 1.0
fs = 100
num_sims = 5 ### display five runs
t_init = 0
t_end  = 700
n = 1000### Compute 1000 grid points
dt     = math.pi *2*min(omega, w)/n
x_init = 0
z_init = 1.0
ts    = np.arange(t_init, t_end, dt)
N = ts.size

def mu(x, z, t):
    return -(gamma*z + omega*x)
    
def Cos(t): 
    return A*math.cos(w*t)

xs    = np.zeros(N)
zs    = np.zeros(N)

xs[0] = x_init
zs[0] = z_init

for i in range(1, ts.size):
    t = (i-1) * dt
    x = xs[i-1]
    z = zs[i-1]
    zs[i] = z + mu(x, z, t) * dt + Cos(t)
    xs[i] = x + z*dt
NN = 1000
time = np.arange(NN) / fs
freqs = np.fft.fftfreq(time.size, 1/fs)
idx = np.argsort(freqs)
ps = np.abs(np.fft.fft(xs))
plt.figure()
plt.loglog(freqs[idx], ps[idx])
plt.title('Power spectrum (np.fft.fft)')
```
