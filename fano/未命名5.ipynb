```python
import numpy as np
import matplotlib.pyplot as plt
import math

omega = 1.0
gamma = 0.06
A = 1.0
w = 0.9

num_sims = 5 ### display five runs
t_init = 0
t_end  = 700
n = 1000### Compute 1000 grid points
dt     = math.pi *2/(n*min(omega,w))
x_init = 0
z_init = 1.0
ts    = np.arange(t_init, t_end, dt)
N = ts.size

def mu(x, z, t):
    return -(gamma*z + omega*x)
    
def Cos(t): 
    return A*math.cos(w*t)

xs    = np.zeros(N)
zs    = np.zeros(N)

xs[0] = x_init
zs[0] = z_init

for i in range(1, ts.size):
    t = (i-1) * dt
    x = xs[i-1]
    z = zs[i-1]
    zs[i] = z + mu(x, z, t) * dt + Cos(t)
    xs[i] = x + z*dt
plt.plot(ts, xs)
plt.show()
```
